#**¿Que son las UVI, las UVA y dólar blue ?**

  * El dólar blue es la divisa que se compra y se vende en el mercado informal practicamente en forma inmediata.  

  * UVI,unidades de cuenta para invertir, ahorrar o tomar préstamos. Son unidades de cuenta porque están relacionadas con el precio de otros bienes y servicios: 1000 UVIs valen lo mismo que un metro cuadrado construido.

  *UVA, es la Unidad de Valor Adquisitivo actualizable por "CER", valor que se actualiza diariamente y lo podés consultar en la página del Banco Central de la República Argentina.



#**¿Que es un plazo fijo y que significa que sea  precancelable?**

En la mayoria de los casos es [precancelable](https://www.bcra.gob.ar/bcrayvos/Preg-Frec-Se-puede-cancelar-anticipadamente-un-Plazo-Fijo.asp) al cumplirse los 30  dias de corrido. La posibilidad de cancelarlo antes de tiempo suele tardar unos dias en algunos casos hasta 5 dias habiles.

# TODO: Datos entre dos fechas 

```python
Fecha1='2022-02-04'
Fecha2='2022-07-16'
Fecha3='2022-09-01'

def datos_entre(Fecha1,Fecha2):
  indices_1_2 = indices_df[(indices_df['Fecha']> Fecha1) & (indices_df['Fecha']< Fecha2)].copy()
  indices_1_2['UVA_index'] = indices_1_2['UVA_index'] / (indices_1_2.iloc[0]['UVA_index']) * 100
  indices_1_2['UVI_index'] = indices_1_2['UVI_index'] / (indices_1_2.iloc[0]['UVI_index']) * 100
  indices_1_2['i_USD'] = indices_1_2['i_USD'] / (indices_1_2.iloc[0]['i_USD']) * 100
  return indices_1_2

def grafico_entre(Fecha1,Fecha2):
  df=datos_entre(Fecha1,Fecha2)
  fig = px.line(df, x="Fecha", y=["UVA_index","UVI_index","i_USD"], title='Indice de precios Construcción (UVI) vs Consumidor (UVA) vs Dolar Blue')
  fig.show()
    
```

# **¿Que más podemos aprender del gráfico?**
  De los gráficos a continuación podemos concluir que en el primero, se observa que la mejor inversion es la del plazo fijo  UVA por un peridodo bastante prolongado estan muy por encima de las demás.El el segundo  solo por un pequeño periodo el dólar blue se disparo pero luego volvio a sus valores en menos de 15 dias. 

¿Que conlcusión nos deja el ultimo grafico?
Muestra que no necesariamente por un incremento en la cotizacion o en rendimiento de una inversion debemos cambiarnos. Solo deberiamos cambiar de cuando tenemos algùn tipo de certeza en que se puede llegar a tener una corriente alsista y esta puede ser estable.    



```python

grafico_entre(Fecha1,Fecha2)
grafico_entre(Fecha2,Fecha3)
```
